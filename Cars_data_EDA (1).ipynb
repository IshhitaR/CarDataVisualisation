```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='make', y='ex-showroom_price', data=df.head(100), hue='fuel_type', split=True, inner='quartile', palette='muted')
plt.xlabel('Make')
plt.ylabel('Ex-showroom Price')
plt.title('Violin Plot of ex-showroom price by make and fuel type')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='fuel_type', data= df, palette='Set3')
plt.title("Fuel Type Distribution")
plt.show()
```

```python
figsize=(8, 1.2 * len(df['fuel_type'].unique()))
plt.figure(figsize = figsize)
sns.violinplot(df, x='cylinders', y='fuel_type', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
top_make_counts = df.model.value_counts().head(10)
plt.figure(figsize=(8,6))
sns.barplot(x=top_make_counts.values, y=top_make_counts.index, palette='viridis')
plt.xlabel('Count')
plt.ylabel('Car Model')
plt.title("Top 10 Car Model Distribution")
plt.show()
```

```python
fig = plt.figure(figsize = (10,6))
ax = fig.subplots()
df.make.value_counts().head(20).plot(ax=ax, kind='pie')
ax.set_ylabel("")
ax.set_title("Top Car making companies in India")
plt.show()
```

```python
sns.countplot(data=df, x="make", palette="CMRmap")
plt.xticks(rotation = 90)
plt.xlabel("Make", fontsize=10, color='black')
plt.ylabel("Count", fontsize=10, color='black')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, y='body_type', alpha=.6, palette='plasma')
plt.title('CArs by Car body type', fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('')
plt.ylabel('')
```

```python
fig= px.scatter_3d(df, x='displacement', z='ex-showroom_price', y='fuel_type', color='make', width=800, height=750)
fig.update_layout(showlegend=True)
fig.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['number_of_airbags'], palette='husl')
plt.xlabel("number of airbags")
plt.ylabel('count')
plt.title('Distribution of Number of Airbags')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
